```python
import pandas as pd

df = pd.read_csv("/workspaces/Data_3500_notes/Lecture code/NVIDIA CORPORATION (04-16-2024 _ 11-05-2025).csv")

```

```python
#clean the string

df["Close/Last"]= df["Close/Last"].replace("[$]","", regex = True).astype(Float)
df.info()

df = df.iloc(::-1).reset_index(drop = True)
df


```

```python
#moving average

df["Moving Avg"] = df["Close/Last"].shift(1).rolling(window = 5).mean()
df.head(10)
```

```python
buy = 0
first_buy = 0
total_profit = 0

for i in range(len(df)):
    current_price = df.loc[i, "Close/Last"]
    average_price = df.loc[i, "Moving Avg"]

    if current_price < average_price *.98:
        buy = current_price
        if first_buy == current_price:
            print("Buying at ", buy)
    elif current_price > (average_price) *1.02:
        total_profit += current_price-buy
        trade_profit = buy - current_price
        print("Selling at: ", current_price)
        print("Trade Profit: ", trade_profit)
```
