Programming Activity 3

Using the Titanic dataset and the Pandas library, find the average survival rate by class and sex.

```python
survive_class_sex = df.groupby(["Pclass","Sex"])["Survived"].mean()

print("Average survival rate by class and sex:", survive_class_sex)
```

Programming Activity 6
Using the weather file (located on week 11 programming activities page), find the longest warm streak where the high temperature was above 65 degrees.

```python
import pandas as pd

df = pd.read_csv("/workspaces/Data_3500_notes/Lecture code/SLC_weather.csv")
```

```python
df["Hot"] = df["Max Temp"] > 65

current_streak = 0
longest_streak = 0

for is_hot in df["Hot"]:
    if is_hot:
        current_streak += 1
        if current_streak > longest_streak:
            longest_streak = current_streak
        else:
            current_streak = 0

print("Longest streak:", longest_streak, "days")

```

Programming Activity 7

Using the same weather dataset, find the hottest day and the coldest night.

```python


hottest_day = df.loc[df["Max Temp"].idxmax()]
coldest_night = df.loc[df["Min Temp"].idxmin()]

print("Coldest night:", coldest_night)
print("Hottest day:", hottest_day)
```

Programming Activity 8
Using the same weather dataset, find the single most volatile day (the day with the biggest difference between high and low temperature).


```python
df["Volatility"] = df["Max Temp"] - df["Min Temp"]

most_volatile = df.loc[df["Volatility"].idxmax()]

print("Most volatile day:", most_volatile)
```
